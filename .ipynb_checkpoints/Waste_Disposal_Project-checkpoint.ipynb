```python
# ============================================================
# EPA LICENSEE WASTE DATA (2021-2024) - COMPLETE PIPELINE
# ============================================================

# 1. IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import glob
import os
import re

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

# ============================================================
# 2. DEFINE FOLDER PATH
# ============================================================

folder_path = r"C:\Users\55929\Desktop\CCT_Colege\dataset"

# Find all Excel files in the folder
files = glob.glob(os.path.join(folder_path, "*.xlsx"))

print("Files found:")
for f in files:
    print(f)
```

```python
# ============================================================
# 3. LOAD ONLY FILES FROM 2021 TO 2024
# ============================================================

df_list = []

for file in files:
    filename = os.path.basename(file)

    # Extract year from file name
    match = re.search(r'20(21|22|23|24)', filename)
    if match:
        year = int(match.group())
        temp_df = pd.read_excel(file)
        temp_df["year"] = year
        df_list.append(temp_df)

if len(df_list) == 0:
    raise ValueError("No Excel files from 2021 to 2024 were found.")

# Concatenate all files
df = pd.concat(df_list, ignore_index=True)
```

```python
print("\nCombined dataset shape:")
print(df.shape)
```

```python
print("\nFirst 5 rows:")
df.head()
```

```python
# ============================================================
# 4. STANDARDIZE COLUMN NAMES
# ============================================================

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("/", "_", regex=False)
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
)

print("\nStandardized columns:")
df.columns.tolist()
```

```python
# ============================================================
# 5. CONVERT DATA TYPES
# ============================================================

numeric_cols = [
    "accepted_waste_quantity",
    "accepted_waste_quantity_from_abroad",
    "accepted_waste_quantity_from_ireland",
    "waste_quantity_tonnes_per_year"
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

yes_no_cols = [
    "final_destination_also_next_destination",
    "next_destination_abroad",
    "final_destination_abroad"
]

for col in yes_no_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.title()

# Keep only valid rows for the main continuous variable
if "accepted_waste_quantity" in df.columns:
    df = df[df["accepted_waste_quantity"].notna()]
    df = df[df["accepted_waste_quantity"] >= 0]
```

```python
print("\nDataset shape after cleaning:")
print(df.shape)
```

```python
print("\nMissing values:")
print(df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ============================================================
# 1. Selecionar apenas colunas numéricas
# ============================================================

numeric_df = df.select_dtypes(include=['int64', 'float64'])

print("Colunas numéricas usadas na correlação:")
print(numeric_df.columns)

# ============================================================
# 2. Calcular matriz de correlação
# ============================================================

correlation_matrix = numeric_df.corr()

# ============================================================
# 3. Plotar Heatmap
# ============================================================

plt.figure(figsize=(10, 8))

sns.heatmap(
    correlation_matrix,
    annot=True,          # mostra valores
    fmt=".2f",           # 2 casas decimais
    cmap="coolwarm",     # cores
    linewidths=0.5
)

plt.title("Heatmap of Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
# remover colunas com muitos valores ausentes
numeric_df = numeric_df.dropna(axis=1, thresh=int(0.7 * len(numeric_df)))

correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Filtered Correlation Heatmap")
plt.show()
```

```python
# ============================================================
# 6. CREATE DERIVED VARIABLES
# ============================================================

if "next_destination_abroad" in df.columns:
    df["export_next_binary"] = df["next_destination_abroad"].map({"Yes": 1, "No": 0})

if "final_destination_abroad" in df.columns:
    df["export_final_binary"] = df["final_destination_abroad"].map({"Yes": 1, "No": 0})

if "accepted_waste_quantity" in df.columns:
    df["log_accepted_waste_quantity"] = np.log1p(df["accepted_waste_quantity"])

# Count records per facility and year
if "licence_profile" in df.columns:
    records_per_facility_year = (
        df.groupby(["year", "licence_profile"])
          .size()
          .reset_index(name="record_count")
    )

    unique_codes_per_facility_year = (
        df.groupby(["year", "licence_profile"])["list_of_waste_code"]
          .nunique()
          .reset_index(name="unique_waste_codes_count")
    )

```

```python
# ============================================================
# 7. DESCRIPTIVE STATISTICS
# ============================================================

print("\n================ OVERALL DESCRIPTIVE STATISTICS ================")
existing_numeric = [col for col in numeric_cols if col in df.columns]
print(df[existing_numeric].describe())

print("\n================ DESCRIPTIVE STATISTICS BY YEAR ================")
if "accepted_waste_quantity" in df.columns:
    yearly_summary = (
        df.groupby("year")["accepted_waste_quantity"]
          .agg(["count", "mean", "median", "std", "min", "max"])
    )
    print(yearly_summary)

# Categorical summaries
cat_cols = [
    "waste_source",
    "waste_action",
    "list_of_waste_code",
    "on_site_waste_treatment_type",
    "next_destination_abroad",
    "final_destination_abroad"
]

for col in cat_cols:
    if col in df.columns:
        print(f"\nTop categories for {col}:")
        print(df[col].value_counts(dropna=False).head(10))

```

```python
# ============================================================
# 8. VISUALIZATIONS (AT LEAST 3)
# ============================================================

# GRAPH 1: Histogram of accepted waste quantity
if "accepted_waste_quantity" in df.columns:
    plt.figure(figsize=(8,5))
    plt.hist(df["accepted_waste_quantity"].dropna(), bins=40, edgecolor='black')
    plt.title("Histogram of Accepted Waste Quantity (2021-2024)")
    plt.xlabel("Accepted Waste Quantity (tonnes)")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

```

```python
# GRAPH 2: Histogram of log transformed quantity
if "log_accepted_waste_quantity" in df.columns:
    plt.figure(figsize=(8,5))
    plt.hist(df["log_accepted_waste_quantity"].dropna(), bins=40, edgecolor='black')
    plt.title("Histogram of Log(Accepted Waste Quantity + 1) (2021-2024)")
    plt.xlabel("log(1 + Accepted Waste Quantity)")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

```

```python
# GRAPH 3: Mean accepted waste quantity by year
if "accepted_waste_quantity" in df.columns:
    mean_by_year = df.groupby("year")["accepted_waste_quantity"].mean()
    plt.figure(figsize=(8,5))
    plt.plot(mean_by_year.index, mean_by_year.values, marker='o')
    plt.title("Mean Accepted Waste Quantity by Year")
    plt.xlabel("Year")
    plt.ylabel("Mean Accepted Waste Quantity")
    plt.xticks(mean_by_year.index)
    plt.tight_layout()
    plt.show()

```

```python
# GRAPH 4: Number of records by year
records_by_year = df["year"].value_counts().sort_index()
plt.figure(figsize=(8,5))
plt.bar(records_by_year.index.astype(str), records_by_year.values)
plt.title("Number of Records by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# GRAPH 5: Export proportion by year
if "export_next_binary" in df.columns:
    export_rate_by_year = df.groupby("year")["export_next_binary"].mean()
    plt.figure(figsize=(8,5))
    plt.plot(export_rate_by_year.index, export_rate_by_year.values, marker='o')
    plt.title("Proportion of Next Destination Abroad by Year")
    plt.xlabel("Year")
    plt.ylabel("Proportion")
    plt.xticks(export_rate_by_year.index)
    plt.tight_layout()
    plt.show()

```

```python
# GRAPH 6: Top 10 waste actions
if "waste_action" in df.columns:
    top_actions = df["waste_action"].value_counts().head(10)
    plt.figure(figsize=(10,5))
    plt.bar(top_actions.index.astype(str), top_actions.values)
    plt.title("Top 10 Waste Actions (2021-2024)")
    plt.xlabel("Waste Action")
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

```

```python
# ============================================================
# 9. BINOMIAL DISTRIBUTION ANALYSIS
# ============================================================

print("\n================ BINOMIAL ANALYSIS ================")

if "export_next_binary" in df.columns:
    binomial_data = df["export_next_binary"].dropna()

    n = len(binomial_data)
    x = int(binomial_data.sum())
    p_hat = x / n

    print(f"Overall sample size n = {n}")
    print(f"Overall successes x = {x}")
    print(f"Estimated p_hat = {p_hat:.4f}")
    print(f"Expected value np = {n * p_hat:.2f}")
    print(f"Variance np(1-p) = {n * p_hat * (1 - p_hat):.2f}")

    # By year
    print("\nBinomial summary by year:")
    binomial_yearly = (
        df.groupby("year")["export_next_binary"]
          .agg(["count", "sum", "mean"])
          .rename(columns={"count": "n", "sum": "x", "mean": "p_hat"})
    )
    print(binomial_yearly)

    # Simulated binomial
    simulated_binomial = np.random.binomial(n=50, p=p_hat, size=5000)

    plt.figure(figsize=(8,5))
    plt.hist(simulated_binomial, bins=20, edgecolor='black', density=True, alpha=0.7)
    plt.title("Simulated Binomial Distribution (n=50, p=p_hat)")
    plt.xlabel("Number of exported records in 50 trials")
    plt.ylabel("Density")
    plt.tight_layout()
    plt.show()

    # Normal approximation
    mean_bin = 50 * p_hat
    sd_bin = np.sqrt(50 * p_hat * (1 - p_hat))
    x_line = np.linspace(simulated_binomial.min(), simulated_binomial.max(), 300)

    plt.figure(figsize=(8,5))
    plt.hist(simulated_binomial, bins=20, density=True, edgecolor='black', alpha=0.7)
    plt.plot(x_line, stats.norm.pdf(x_line, mean_bin, sd_bin), linewidth=2)
    plt.title("Binomial Distribution and Normal Approximation")
    plt.xlabel("Count")
    plt.ylabel("Density")
    plt.tight_layout()
    plt.show()

```

```python
# ============================================================
# 10. POISSON DISTRIBUTION ANALYSIS
# ============================================================

print("\n================ POISSON ANALYSIS ================")

if "record_count" in records_per_facility_year.columns:
    poisson_sample = records_per_facility_year["record_count"]

    lambda_hat = poisson_sample.mean()
    sample_variance = poisson_sample.var(ddof=1)

    print(f"Overall lambda_hat = {lambda_hat:.4f}")
    print(f"Overall sample variance = {sample_variance:.4f}")
    print("For Poisson, mean and variance should be approximately similar.")

    # By year
    print("\nPoisson summary by year:")
    poisson_yearly = (
        records_per_facility_year.groupby("year")["record_count"]
        .agg(["count", "mean", "var", "min", "max"])
    )
    print(poisson_yearly)

    # Observed vs Poisson
    plt.figure(figsize=(8,5))
    plt.hist(poisson_sample, bins=20, density=True, edgecolor='black', alpha=0.7)

    x_vals = np.arange(0, poisson_sample.max() + 1)
    pmf_vals = stats.poisson.pmf(x_vals, mu=lambda_hat)
    plt.plot(x_vals, pmf_vals, 'o-', linewidth=2)

    plt.title("Observed Counts per Facility-Year vs Poisson Distribution")
    plt.xlabel("Record count per facility-year")
    plt.ylabel("Density / Probability")
    plt.tight_layout()
    plt.show()

    # Large lambda -> normal approximation
    if lambda_hat > 10:
        poisson_sim = np.random.poisson(lam=lambda_hat, size=5000)
        normal_sim = np.random.normal(loc=lambda_hat, scale=np.sqrt(lambda_hat), size=5000)

        plt.figure(figsize=(8,5))
        plt.hist(poisson_sim, bins=30, density=True, edgecolor='black', alpha=0.6, label='Poisson')
        plt.hist(normal_sim, bins=30, density=True, edgecolor='black', alpha=0.4, label='Normal approx')
        plt.title("Poisson vs Normal Approximation for Large Lambda")
        plt.xlabel("Count")
        plt.ylabel("Density")
        plt.legend()
        plt.tight_layout()
        plt.show()

```

```python
# ============================================================
# 11. NORMAL DISTRIBUTION ANALYSIS
# ============================================================

print("\n================ NORMAL DISTRIBUTION ANALYSIS ================")

if "accepted_waste_quantity" in df.columns:
    original = df["accepted_waste_quantity"].dropna()
    logged = df["log_accepted_waste_quantity"].dropna()

    print("Original Accepted Waste Quantity:")
    print(original.describe())

    print("\nLog-transformed Accepted Waste Quantity:")
    print(logged.describe())

    # QQ plot original
    plt.figure(figsize=(6,6))
    stats.probplot(original, dist="norm", plot=plt)
    plt.title("Q-Q Plot: Accepted Waste Quantity (2021-2024)")
    plt.tight_layout()
    plt.show()

    # QQ plot log
    plt.figure(figsize=(6,6))
    stats.probplot(logged, dist="norm", plot=plt)
    plt.title("Q-Q Plot: Log(Accepted Waste Quantity + 1) (2021-2024)")
    plt.tight_layout()
    plt.show()

    # Shapiro test on subset
    sample_for_test = logged.sample(min(5000, len(logged)), random_state=42)
    shapiro_stat, shapiro_p = stats.shapiro(sample_for_test)

    print(f"\nShapiro-Wilk test on log variable:")
    print(f"Statistic = {shapiro_stat:.4f}")
    print(f"p-value = {shapiro_p:.4g}")

    # Confidence interval for mean log quantity
    mean_logged = logged.mean()
    sd_logged = logged.std(ddof=1)
    n_logged = len(logged)
    se_logged = sd_logged / np.sqrt(n_logged)

    ci_low = mean_logged - 1.96 * se_logged
    ci_high = mean_logged + 1.96 * se_logged

    print(f"\n95% confidence interval for mean log quantity: ({ci_low:.4f}, {ci_high:.4f})")

    # Mean log quantity by year
    mean_log_by_year = df.groupby("year")["log_accepted_waste_quantity"].mean()
    plt.figure(figsize=(8,5))
    plt.plot(mean_log_by_year.index, mean_log_by_year.values, marker='o')
    plt.title("Mean Log(Accepted Waste Quantity + 1) by Year")
    plt.xlabel("Year")
    plt.ylabel("Mean Log Quantity")
    plt.xticks(mean_log_by_year.index)
    plt.tight_layout()
    plt.show()

print("\nPipeline completed successfully.")
```
